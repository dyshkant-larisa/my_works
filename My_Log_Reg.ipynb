```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score, precision_recall_curve, plot_confusion_matrix, plot_roc_curve, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(solver='saga', multi_class='ovr',max_iter=5000)
```

```python
penalty=['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0,1,20)
C = np.logspace(0,10,20)

param_grid = {'penalty':penalty, 'l1_ratio':l1_ratio, 'C':C}
```

```python
grid_model = GridSearchCV(model, param_grid=param_grid)
```

```python
grid_model.fit(X_train, y_train)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
plot_confusion_matrix(grid_model,X_test, y_test)
```

```python
print(classification_report(y_test,y_pred));
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
def plot_multiclass_roc(clf, X_test, y_test, n_classes, figsize=(5,5)):
    y_score = clf.decision_function(X_test)

    # создаём пустые структуры
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # один раз вычисляем dummies 
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc для каждого класса
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
plot_multiclass_roc(grid_model, X_test, y_test, n_classes=3, figsize=(10,6))
```
