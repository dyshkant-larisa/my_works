```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1,2,1,projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
ax = fig.add_subplot(1,2,2, projection = '3d')
p = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth = 0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig,ax = plt.subplots(dpi=100)

ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend(loc=(0.7,0.5));
```

```python
fig,ax = plt.subplots(dpi=90)

ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend(loc=(1.05,0.5));
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(12,8))

ax[0].plot(labels,july16_2007)
ax[0].set_title('July 16th, 2007')

ax[1].plot(labels,july16_2020)
ax[1].set_title('July 16th, 2020');
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(labels,july16_2007, lw=2, color="blue")
ax.set_ylabel("2007", fontsize=18, color="blue")

ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(4)

for label in ax.get_yticklabels():
    label.set_color("blue")
plt.yticks(fontsize=15)    
    
ax1 = ax.twinx()
ax1.plot(labels,july16_2020, lw=2, color="red")
ax1.set_ylabel("2020", fontsize=18, color="red")

ax1.spines['right'].set_color('red')
ax1.spines['right'].set_linewidth(4)

for label in ax1.get_yticklabels():
    label.set_color("red")
    
ax.set_title("July 16th Yield Curves");
plt.yticks(fontsize=15);
```
